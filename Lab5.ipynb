```
import json
import pandas as pd
```

```
with open('/content/drive/My Drive/DATA 301/Colab Notebooks/recipes.txt', "r",encoding= 'unicode_escape') as read_file:
    data = json.load(read_file)
df = pd.DataFrame(data)
df = df[['name','ingredients']]

def get_ingredient_names(arr):
  l = []
  for i in range(len(arr)):
    l.append(arr[i]['name'])
  return l

def str_to_arr(str):
  return [str]

df['ingredients'] = df['ingredients'].apply(get_ingredient_names)
df.ingredients.apply(pd.Series)
df = pd.DataFrame([(d, tup.name) for tup in df.itertuples() for d in tup.ingredients])
df = df.rename(columns={0: 'ingredient', 1: 'recipes'})
df['recipes'] = df['recipes'].apply(str_to_arr)
df = df.groupby(['ingredient']).sum()
df
```
